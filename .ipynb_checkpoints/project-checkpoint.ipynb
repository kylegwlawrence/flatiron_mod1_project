```python
#produce a histogram for each int column
#first get a dataframe of only int columns

#create an empty df to hold the int columns
df_int = pd.DataFrame()
#iterate through the columns of the original df
for col in df.columns:
    #if the column type is int then add to the new df
    if df[col].dtype==np.int64:
        df_int[col]=df[col]
    #if any other data type then pass to the next iteration
    else:
        pass
        

df_int.head()
#check it worked

#then set up a object to hold n graphs
#iterate through each column to plot it's histogram
```
